### dropna() (series + dataframe)

```python
# Remove rows contains null value coresponding to 'name' and 'college'
students.dropna(subset = ['name','college'])
```

```python
# how='all' means a row or column will be dropped only if all its values are Nan.
students.dropna(how='all')
```

```python
# inplace=True ---> Permanant change in DataFrame
students.dropna(inplace=True)
```

### fillna()

```python
students = pd.DataFrame(
    {
        'name':['nitish','ankit','rupesh',np.nan,'partho',np.nan,'abhik',np.nan,'aditya',np.nan],
        'college':['bit','iit','vit',np.nan,np.nan,'vlsi','ssit',np.nan,np.nan,'git'],
        'branch':['eee','it','cse',np.nan,'me','ce','civ','cse','bio',np.nan],
        'cgpa':[6.66,8.25,6.41,np.nan,5.6,9.0,7.4,10,7.4,np.nan],
        'package':[4,5,6,np.nan,6,7,8,9,np.nan,np.nan]
    }
)
```

```python
stud = students.copy()
stud.fillna('---')
```

```python
# Specific columns null value fill up with given number/string
students['name'].fillna('unknown')
```

```python
# Fill null values of cgpa colunm with mean of the cgpa
students['cgpa'].fillna(students['cgpa'].mean())
```

```python
# Fill the City NaN values where Venue is 'Dubai International Cricket Stadium'
ipl.loc[ipl['Venue'] == 'Dubai International Cricket Stadium', 'City']=ipl['City'].fillna('Dubai')
```

```python
# Fill the City NaN values where Venue is 'Sharjah Cricket Stadium'
ipl.loc[ipl['Venue'] == 'Sharjah Cricket Stadium', 'City'] = ipl['City'].fillna('Sharjah')
```
