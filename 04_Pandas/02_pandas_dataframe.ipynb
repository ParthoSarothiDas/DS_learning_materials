# Creating DataFrame

### DataFrame from Dictionary

```python
import pandas as pd

data = {
    "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index = ['day1','day2', 'day3'])
df
```

```python
# using dicts

student_dict = {
    'name':['nitish','ankit','rupesh','rishabh','amit','ankita'],
    'iq':[100,90,120,80,0,0],
    'marks':[80,70,100,50,0,0],
    'package':[10,7,14,2,0,0]
}

student = pd.DataFrame(student_dict)
student.set_index('name',inplace=True)
student
```

# Filtering a DataFrame

```python
# Find all the final winners
df = ipl[ipl['MatchNumber'] == 'Final']
df[['Season','WinningTeam']]
```

```python
# How many super over finishes have occured

ipl[ipl['SuperOver'] == 'Y'].shape[0]
```

```python
# How many matches has 'Chennai Super Kings' won in Kolkata

ipl[(ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')].shape[0]
```

```python
# Toss winner is match winner in percentage

num_match_win = ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]
total_match = ipl.shape[0]
percentage = (num_match_win/total_match) * 100
round(percentage,2)
```

```python
# Number of movies with 'imdb_rating' higher than 8 and 'imdb_votes'>10000

movies[(movies['imdb_rating'] > 8) & (movies['imdb_votes']>10000)].shape[0]
```

```python
# Action movies with rating higher than 7.5

mask1 = movies['genres'].str.contains('Action')
mask2 = movies['imdb_rating'] > 7.5

movies[mask1 & mask2]
```
