#### 4. Kernel Density Estimation (KDE) Plot: When to Use & How to Interpret.
A KDE (Kernel Density Estimation) plot is a smoothed version of a histogram, showing the probability density function (PDF) of a dataset.

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
fig,ax = plt.subplots(nrows=2, ncols=1)
sns.kdeplot(data=tips, x='total_bill', ax=ax[0])
sns.kdeplot(data=tips, y='total_bill', ax=ax[1])
plt.plot()
```

```python
fig,ax = plt.subplots(nrows=1, ncols=2, figsize=(10,6))
sns.kdeplot(data=tips, x='total_bill', bw_adjust=.2, ax=ax[0])
sns.kdeplot(data=tips, x='total_bill', bw_adjust=5, cut=0, ax=ax[1])
ax[0].set_title('Less Smoothing')
ax[1].set_title('More Smoothing')
plt.tight_layout()
plt.plot()
```

```python
# Plot conditional distributions with hue mapping of a second variable:
```

```python
sns.kdeplot(data=tips, x='total_bill', hue='time')
plt.show()
```

```python
# Estimate the cumulative distribution function(s), normalizing each subset.
```

```python
sns.kdeplot(
    data=tips, x="total_bill", hue="time",
    cumulative=True, common_norm=False, common_grid=True,
)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
penguins = sns.load_dataset("penguins")

# Create the countplot
plt.figure(figsize=(8, 5))
ax = sns.countplot(data=penguins, x='species')

# Calculate total count
total = len(penguins)

# Add percentage labels
for p in ax.patches:
    count = p.get_height()
    percentage = f'{(count / total * 100):.1f}%'  # Convert to percentage
    x = p.get_x() + p.get_width() / 2  # Center the text
    y = p.get_height()  # Position above the bar
    ax.annotate(percentage, (x, y), ha='center', va='bottom', fontsize=12, fontweight='bold')

# Show plot
plt.title("Penguin Species Count with Percentage Labels")
plt.ylabel("Count")
plt.xlabel("Species")
plt.show()
```

### 11. jointplot: When to Use & How to Interpret
A jointplot in Seaborn is a powerful combo plot that shows the relationship between two continuous variables, along with their individual distributions. It's like a scatter plot + two histograms/KDEs in one clean visualization.

When to Use jointplot:
 Use when you want to:
    1. Visualize the relationship between two numerical variables.
    2. Simultaneously analyze joint distribution (bivariate) and marginal distributions (univariate).
    3. Detect correlation, clusters, or outliers.
    4. Get a quick overview of a numeric variable pair.

Best for Exploratory Data Analysis (EDA) when:
    1. You're exploring numeric features like height vs. weight, income vs. spending.
    2. You want a one-plot summary showing correlation + histograms/KDE.

```python
penguins.head()
```

```python
sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', kind='reg')
plt.show()
```

```python
# Simple jointplot
sns.jointplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
# Draw a jointplot with regression line to to assess linear trend
sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg')
plt.show()
```

```python
# Draw a jointplot with kde to  smoothed density estimates.
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde')
plt.show()
```

```python
# Draw a jointplot with hist to dealing with large datasets.
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hist')
plt.show()
```

```python
# Draw a jointplot with Hexbin plot to dealing with large datasets.
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
plt.show()
```

```python
# Draw a jointplot with Residual plot.For checking model residuals
sns.jointplot(data=tips, x='total_bill', y='tip', kind='resid')
plt.show()
```

```python
# Draw a jointplot with hue
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```
