### Matplotlib Visualization - Basics

***Introduction:***
Matplotlib is one of the most widely used Python libraries for data visualization. It provides a flexible platform for creating a wide range of static, animated, and interactive plots in Python.

Library name   : matplotlib

Main module    : pyplot (usually imported as plt)

Install via pip: pip install matplotlib

#### Matplotlib Plot Types
| Plot Type       | Function            | Description                                                                 | Typical Use Case                                                 |
|------------------|---------------------|-----------------------------------------------------------------------------|------------------------------------------------------------------|
| Line Plot        | `plt.plot()`        | Plots data as a line connecting points.                                     | Trends over time, time-series data                              |
| Scatter Plot     | `plt.scatter()`     | Plots individual data points without connecting lines.                      | Relationship between two variables, clustering                   |
| Bar Chart        | `plt.bar()`         | Vertical bars representing categorical data.                                | Comparing values across categories                              |
| Horizontal Bar   | `plt.barh()`        | Horizontal version of bar chart.                                            | Long category names, better horizontal space utilization         |
| Histogram        | `plt.hist()`        | Groups data into bins and shows frequency.                                  | Distribution of numerical data                                  |
| Box Plot         | `plt.boxplot()`     | Shows distribution using quartiles and outliers.                            | Statistical summary, detecting outliers                         |
| Pie Chart        | `plt.pie()`         | Circular chart showing proportions.                                         | Showing parts of a whole                                         |
| Area Plot        | `plt.stackplot()`   | Stacked areas representing quantities over time.                            | Cumulative data trends over time                                |
| Stem Plot        | `plt.stem()`        | Plots data points as stems from a baseline.                                 | Discrete data visualization                                     |
| Step Plot        | `plt.step()`        | Plots values with steps, rather than a continuous line.                     | Digital signals, piecewise data                                 |
| Error Bar Plot   | `plt.errorbar()`    | Adds error bars to line or scatter plots.                                   | Displaying uncertainty in measurements                          |
| Fill Between     | `plt.fill_between()`| Shades the area between two curves.                                         | Highlighting range, confidence intervals                        |
| Contour Plot     | `plt.contour()`     | Shows 3D surface levels in 2D using contour lines.                          | Topography, optimization functions                              |
| Filled Contour   | `plt.contourf()`    | Contour plot with filled colors between levels.                             | Same as above, with more emphasis on areas                      |
| Image Plot       | `plt.imshow()`      | Displays image data or heatmaps.                                            | Visualizing matrices, images, heatmaps                          |
| Heatmap (Alt)    | `plt.pcolormesh()`  | 2D heatmap from a grid of values.                                           | Similar to imshow but more customizable                         |
| Polar Plot       | `plt.polar()`       | Plots data in polar coordinates (theta, r).                                 | Wind direction, periodic data                                   |
| 3D Plot (Line)   | `ax.plot3D()`       | 3D line plot using `mpl_toolkits.mplot3d`.                                  | 3D data visualization                                           |
| 3D Scatter       | `ax.scatter3D()`    | 3D scatter plot.                                                            | 3D point cloud data                                             |
| 3D Surface Plot  | `ax.plot_surface()` | 3D surface mesh using x, y, z values.                                       | 3D continuous surfaces, mathematical surfaces                   |


#### Matplotlib Plot Customization Options
| Customization         | Parameter / Function        | Typical Use Case                                           |
|------------------------|-----------------------------|------------------------------------------------------------|
| Title                  | `plt.title('Title')`        | Add a title to the plot                                    |
| X-axis Label           | `plt.xlabel('X Label')`     | Label the X-axis                                           |
| Y-axis Label           | `plt.ylabel('Y Label')`     | Label the Y-axis                                           |
| Legend                 | `plt.legend()`              | Display labels for different plot lines                    |
| Line Color             | `color='red'` or `'r'`      | Change color of lines/markers                              |
| Line Style             | `linestyle='--'`            | Dashed, dotted, solid lines                                |
| Line Width             | `linewidth=2`               | Thicker or thinner lines                                   |
| Marker Style           | `marker='o'`, `'^'`, `'*'`  | Custom point markers                                       |
| Marker Size            | `markersize=8`              | Adjust size of data point markers                          |
| Grid                   | `plt.grid(True)`            | Add gridlines to the plot                                  |
| Axis Limits            | `plt.xlim()`, `plt.ylim()`  | Set custom ranges for X and Y axes                         |
| Tick Marks             | `plt.xticks()`, `plt.yticks()`| Customize tick positions and labels                      |
| Tick Rotation          | `rotation=45` in xticks     | Rotate tick labels for readability                         |
| Figure Size            | `plt.figure(figsize=(w,h))` | Set plot canvas size (width, height)                       |
| Background Color       | `facecolor='lightgray'`     | Change figure or axes background color                     |
| Annotation             | `plt.annotate()`            | Add labels with arrows on specific data points             |
| Alpha (Transparency)   | `alpha=0.6`                 | Set transparency level for plot elements                   |
| Font Size              | `fontsize=14`               | Adjust size of title/labels/ticks                          |
| Subplot Spacing        | `plt.tight_layout()`        | Prevent overlap of subplots                                |
| Save Figure            | `plt.savefig('file.png')`   | Save the current plot as image                             |
| Axis Aspect Ratio      | `plt.axis('equal')`         | Make the scale on x and y axis equal                       |
| Adding Text            | `plt.text(x, y, 'text')`    | Add custom text at specified location                      |
| Fill Between Curves    | `plt.fill_between()`        | Shade area between curves                                  |
| Multiple Plots Layout  | `plt.subplot()`             | Create multi-panel plots in a single figure                |


 ***Notes:***
   - Use plt.figure() before plotting if you want to define figure-level properties like size or background color.
   - Combining multiple customizations results in clean and readable plots.
   - Most styling parameters are available directly in plt.plot() or similar plotting functions.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Type of Plots

#### 1. Line Plot â€“ plt.plot()

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Line Plot")
plt.show()
```

#### 2. Scatter Plot â€“ plt.scatter()

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

#### 3. Bar Chart â€“ plt.bar()

```python
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

plt.bar(categories, values)
plt.title("Bar Chart")
plt.show()
```

#### 4. Horizontal Bar Chart â€“ plt.barh()

```python
plt.barh(categories, values)
plt.title("Horizontal Bar Chart")
plt.show()
```

#### 5. Histogram â€“ plt.hist()

```python
data = np.random.randn(1000)

plt.hist(data, bins=30)
plt.title("Histogram")
plt.show()
```

#### 6. Box Plot â€“ plt.boxplot()

```python
data = [np.random.randn(100), np.random.randn(100) + 2]

plt.boxplot(data)
plt.title("Box Plot")
plt.show()
```

#### 7. Pie Chart â€“ plt.pie()

```python
sizes = [25, 35, 20, 20]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

#### 8. Area Plot â€“ plt.stackplot()

```python
x = np.arange(1, 6)
y1 = [1, 3, 4, 5, 2]
y2 = [2, 2, 3, 4, 1]

plt.stackplot(x, y1, y2, labels=['y1', 'y2'])
plt.legend()
plt.title("Area Plot")
plt.show()
```

#### 9. Stem Plot â€“ plt.stem()

```python
x = np.linspace(0, 2 * np.pi, 10)
y = np.cos(x)

plt.stem(x, y)
plt.title("Stem Plot")
plt.show()
```

#### 10. Step Plot â€“ plt.step()

```python
x = np.arange(10)
y = np.sin(x)

plt.step(x, y, where='mid')
plt.title("Step Plot")
plt.show()
```

#### 11. Error Bar Plot â€“ plt.errorbar()

```python
x = np.linspace(0.1, 4, 10)
y = np.exp(-x)
error = 0.1 * y

plt.errorbar(x, y, yerr=error, fmt='o')
plt.title("Error Bar Plot")
plt.show()
```

#### 12. Fill Between â€“ plt.fill_between()

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.fill_between(x, y1, y2, color='gray', alpha=0.3)
plt.title("Fill Between")
plt.show()
```

#### 13. Contour Plot â€“ plt.contour()

```python
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2)

plt.contour(X, Y, Z)
plt.title("Contour Plot")
plt.show()
```

#### 14. Filled Contour Plot â€“ plt.contourf()

```python
plt.contourf(X, Y, Z, cmap='viridis')
plt.title("Filled Contour Plot")
plt.colorbar()
plt.show()
```

#### 15. Image Plot â€“ plt.imshow()

```python
data = np.random.rand(10, 10)

plt.imshow(data, cmap='hot', interpolation='nearest')
plt.title("Image Plot")
plt.colorbar()
plt.show()
```

#### 16. Heatmap (Alt) â€“ plt.pcolormesh()

```python
x = np.linspace(0, 1, 11)
y = np.linspace(0, 1, 11)
X, Y = np.meshgrid(x, y)
Z = np.sin(10 * X) * np.cos(10 * Y)

plt.pcolormesh(X, Y, Z, shading='auto', cmap='viridis')
plt.title("Heatmap using pcolormesh")
plt.colorbar()
plt.show()
```

#### 17. Polar Plot â€“ plt.polar()

```python
theta = np.linspace(0, 2 * np.pi, 100)
r = 1 + np.sin(5 * theta)

plt.polar(theta, r)
plt.title("Polar Plot")
plt.show()
```

#### 18. 3D Line Plot â€“ ax.plot3D()

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.sin(x)
z = np.cos(x)

ax.plot3D(x, y, z)
ax.set_title("3D Line Plot")
plt.show()
```

#### 19. 3D Scatter Plot â€“ ax.scatter3D()

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

ax.scatter3D(x, y, z, color='green')
ax.set_title("3D Scatter Plot")
plt.show()
```

#### 20. 3D Surface Plot â€“ ax.plot_surface()

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-3, 3, 50)
y = np.linspace(-3, 3, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("3D Surface Plot")
plt.show()
```

### subplot()

#### 1. Using plt.subplot(nrows, ncols, index)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2, 1, 1)  # (rows=2, cols=1, first plot)
plt.plot(x, y1)
plt.title("Sine")

plt.subplot(2, 1, 2)  # (rows=2, cols=1, second plot)
plt.plot(x, y2)
plt.title("Cosine")


plt.tight_layout()
plt.show()
```

#### 2. Using plt.subplots()
This method returns a figure and axes object, and is more flexible for looping, labeling, etc.

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 6))

x = np.linspace(0, 10, 100)
axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("sin(x)")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("cos(x)")

axs[1, 0].plot(x, np.tan(x))
axs[1, 0].set_title("tan(x)")

axs[1, 1].plot(x, np.exp(-x))
axs[1, 1].set_title("exp(-x)")

plt.tight_layout()
plt.show()
```

#### 3. Using fig.add_subplot()
Useful in OOP-style plotting where you manually manage figure and axes.

```python
fig = plt.figure(figsize=(8, 4))

ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(x, np.sin(x))
ax1.set_title("Sine")

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(x, np.cos(x))
ax2.set_title("Cosine")

plt.tight_layout()
plt.show()
```
